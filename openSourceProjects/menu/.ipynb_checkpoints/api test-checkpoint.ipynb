```python
geo_data = pd.read_excel("geo.xlsx")
geo_data
```

```python
elevation_list = []

for index, row in geo_data.iterrows():
#     z.append(row.x + row.y)
#     print(index,row)
    lat = str(row.latitude)
    lon = str(row.longitude)
#     print(lat,lon)
    url = elevation_base_url + "?locations=" + lat + "," + lon + key
#     print (url)
    response = simplejson.load(urllib.request.urlopen(url))
    elevation = response['results'][0]['elevation']
    print(elevation)
    elevation_list.append(elevation)
```

```python
geo_data['elevation'] = elevation_list
```
